# Detecting and Filtering Outliers

```python
data1 = pd.DataFrame(np.random.randn(1000, 4))
data1
```

```python
data1[(np.abs(data1) > 3).any(1)]
```

```python
data1.describe()
```

```python
# Keeping data in interval -3 to 3.
data1[(np.abs(data1) > 3).any(1)] = np.sign(data1) * 3
data1
```
