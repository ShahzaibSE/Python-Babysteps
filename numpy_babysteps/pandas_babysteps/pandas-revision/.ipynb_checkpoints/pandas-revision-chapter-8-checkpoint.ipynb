DataFrame’s set_index function will create a new DataFrame using one or more of

```python
frame.set_index(['c', 'd'], drop=False)
```

## 8.2 Combining and Merging Datasets

## Database-Style DataFrame Joins

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],'data1': range(7)})
df1
```

```python
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],'data2': range(3)})
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on='key')
```

```python
# If the column names are different in each object, you can specify them separately.
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],'data1': range(7)})
df3
```

```python
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],'data2': range(3)})
df4
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
df1
```

```python
df2
```

```python
# Outer-join merge.
pd.merge(df1, df2, how='outer')
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, how='left')
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, how='right',indicator=True)
```

```python
# To merge with multiple keys, pass a list of column names.
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],'key2': ['one', 'two', 'one'],'lval': [1, 2, 3]})
left
```

```python
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],'key2': ['one', 'one', 'one', 'two'],
                        'rval': [4, 5, 6, 7]})
right
```

```python
pd.merge(left, right, on=['key1', 'key2'], how='outer')
```

```python
pd.merge(left, right, on='key1')
```

```python
pd.merge(left, right, on='key1', suffixes=['_left','_right'],indicator=True)
```

## Merging on Index

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                        'value': range(6)})
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
```

```python
left1
```

```python
right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, indicator=True)
```

Since the default merge method is to intersect the join keys, you can instead form the
union of them with an outer join.

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

```python
# DataFrame has a convenient join instance for merging by index. 
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],index=['a', 'c', 'e'],columns=['Ohio', 'Nevada'])
left2
```

```python
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],index=['b', 'c', 'd', 'e'],
                                                                    columns=['Missouri', 'Alabama'])
right2
```

```python
sframe = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]])
sframe
```

```python
left2.join(right2, how='outer')
```

```python
left1
```

```python
right1
```

```python
left1.join(right1, on='key') #Default Join left.
```

```python
another = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [16., 17.]],index=['a', 'c', 'e', 'f'],
                       columns=['New York', 'Oregon'])
another
```

```python
left2.join([right2, another])
```

```python
left2.join([right2, another], how='outer')
```

```python
left2.join([right2, another], how='inner')
```

### With Series.

```python
s1 = pd.Series([0, 1], index=['a', 'b'])
s2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'])
s3 = pd.Series([5, 6], index=['f', 'g'])
```

```python
s1
```

```python
s2
```

```python
s3
```

```python
pd.concat([s1,s2,s3])
```

```python
pd.concat([s1, s2, s3], axis=1)
```

```python
pd.concat([s1, s2, s3], axis=1, join='inner')
```

```python
s1
```

```python
s3
```

```python
s4 = pd.concat([s1, s3])
s4
```

```python
pd.concat([s1, s4], axis=1)
```

```python
pd.concat([s1, s4], axis=1, join='inner')
```

```python
# You can even specify the axes to be used on the other axes with join_axes
pd.concat([s1, s4], axis=1, join_axes=[['a', 'c', 'b', 'e']])
```

```python
result = pd.concat([s1, s1, s3], keys=['one', 'two', 'three']) #On axis '0'.
result # Returned Hierachial Index.
```

```python
result.unstack()
```

```python
result.unstack().stack()
```

#### In the case of combining Series along axis=1, the keys become the DataFrame column headers.

```python
pd.concat([s1, s2, s3], axis=1, keys=['one', 'two', 'three']) #keys becomes column.
```

### The same logic extends to DataFrame objects.

```python
df1 = pd.DataFrame(np.arange(6).reshape(3, 2), index=['a', 'b', 'c'],
                    columns=['one', 'two'])
df2 = pd.DataFrame(5 + np.arange(4).reshape(2, 2), index=['a', 'c'],
                    columns=['three', 'four'])
```

```python
df1
```

```python
df2
```

```python
#If you pass a dict of objects instead of a list, the dict’s keys will be used for the keys option.
pd.concat([df1, df2], axis=1, keys=['level1', 'level2'])
```

```python
#In the case of combining Series along axis=1, the keys become the DataFrame column headers.
pd.concat([s1, s2, s3], axis=1, keys=['one', 'two', 'three'])
```

```python
pd.concat({'level1': df1, 'level2': df2}, axis=1)
```

```python
pd.concat([df1, df2], axis=1, keys=['level1', 'level2'],names=['upper', 'lower'])
```

#### A last consideration concerns DataFrames in which the row index does not contain any relevant data.

```python
df1 = pd.DataFrame(np.random.randn(3, 4), columns=['a', 'b', 'c', 'd'])
df2 = pd.DataFrame(np.random.randn(2, 3), columns=['b', 'd', 'a'])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
pd.concat([df1, df2])
```

## Combining Data with Overlap

```python
a = pd.Series([np.nan, 2.5, np.nan, 3.5, 4.5, np.nan],index=['f', 'e', 'd', 'c', 'b', 'a'])
b = pd.Series(np.arange(len(a), dtype=np.float64),index=['f', 'e', 'd', 'c', 'b', 'a'])
```

```python
a
```

```python
b
```

```python
b[-1] = np.nan
b
```

```python
np.where(pd.isnull(a), b, a)
```

```python
# Series has a combine_first method, which performs the equivalent of this operation 
# along with pandas’s usual data alignment logic
b[:-2].combine_first(a[2:])
```

With DataFrames, combine_first does the same thing column by column, so you
can think of it as “patching” missing data in the calling object with data from the
object you pass.

```python
df1 = pd.DataFrame({'a': [1., np.nan, 5., np.nan], 'b': [np.nan, 2., np.nan, 6.],'c': range(2, 18, 4)})
df2 = pd.DataFrame({'a': [5., 4., np.nan, 3., 7.],'b': [np.nan, 3., 4., 6., 8.]})
```

```python
df1
```

```python
df2
```

```python
df1.combine_first(df2)
```

```python
df1.unstack
```

```python
df1.stack
```

## Pivoting “Long” to “Wide” Format

```python
data = pd.read_csv('examples/macrodata.csv')
data
```

```python
data.head()
```

```python
periods = pd.PeriodIndex(year=data.year, quarter=data.quarter,name='date')
periods
```

```python
columns = pd.Index(['realgdp', 'infl', 'unemp'], name='item')
columns
```

```python
data = data.reindex(columns=columns)
data.index = periods.to_timestamp('D', 'end')
ldata = data.stack().reset_index().rename(columns={0: 'value'})
```

```python
ldata[:10]
```

```python
pivoted = ldata.pivot('date', 'item', 'value')
pivoted
```

```python
ldata['value2'] = np.random.randn(len(ldata))
ldata
```

```python
pivoted = ldata.pivot('date', 'item')
pivoted
```

```python
pivoted[:5]
```

## Pivoting “Wide” to “Long” Format

```python
df = pd.DataFrame({'key': ['foo', 'bar', 'baz'],'A': [1, 2, 3],'B': [4, 5, 6],'C': [7, 8, 9]})
df
```

```python
melted = pd.melt(df, ['key'])
# print(melted['key2']) #KeyError: 'key2'
melted
```

```python
reshaped = melted.pivot('key', 'variable', 'value')
reshaped
```

```python
reshaped.reset_index()
```

```python
pd.melt(df, id_vars=['key'], value_vars=['A', 'B'])
```

```python
pd.melt(df, value_vars=['A', 'B', 'C'])
```

```python
pd.melt(df, value_vars=['key', 'A', 'B'])
```
